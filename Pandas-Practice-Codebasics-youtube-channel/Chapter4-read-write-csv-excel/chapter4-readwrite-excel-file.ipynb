```python
#to convert 'not available'/'n.a.' too NAN use below call
df = pd.read_csv('stock_data.csv', na_values=['not available','n.a.'])
df
```

```python
# if we want to give seperate treatment for each column regarding NAN we can use dictionary
df = pd.read_csv('stock_data.csv',na_values= {
    'eps':['not available','n.a.'],
    'revenue':['not available','n.a.', -1],
    'people':['not available','n.a.']
})
df
```

```python
df.to_csv('new.csv', index=False) # index parameter will not export index column
#if we want to export only specific column then we can use below call
df.to_csv('new1.csv', columns=['tickers','eps'], index=False)
```

```python
df = pd.read_excel('stock_data.xlsx','Sheet1')
df
```

```python
# If we want to write converter to any cell then we can write it as

def converter_people_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell

df = pd.read_excel('stock_data.xlsx','Sheet1', converters={
    'people':converter_people_cell
})

df
```

```python
# to write data data to excel sheet we can use below command
df.to_excel('new.xlsx',sheet_name='Sheet1',startcol=2,startrow=3)
```

```python
# If we want to write 2 dataframes as 2 sheets in one excel we can do nelow thing
df_stocks = pd.DataFrame({
    'tickers':['google','wmt'],
    'price':[30,20],
    'pe':[1.1,2.2],
    'eps':[4.4,5.5]
})
               
df_weather = pd.DataFrame({
    'day':['1/1/2017','1/2/2017'],
    'temperature':[23,26],
    'event':['rain','sunny']
})
       
with pd.ExcelWriter('stocks-weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name='stocks')
    df_weather.to_excel(writer, sheet_name='weather')
```
