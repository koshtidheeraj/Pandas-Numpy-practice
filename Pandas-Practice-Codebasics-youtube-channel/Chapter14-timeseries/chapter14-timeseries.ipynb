```python
import pandas as pd

df = pd.read_csv('aapl.csv', parse_dates=['Date'], index_col='Date') # paarse_dates will consider columns in
# given list as date type and index_col makes Date as index df
df.head(4)
```

```python
# we can do resampling . e.g. data whic we have is daily data, we can pick one column and for that
# column we can generate monthly data by using arithmetic operation on values of rows. Let s take mean operation

%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python
# in case we have data with no  date colums and we have to create the date column then
# we can create the date columnc using date_brange

df2 =  pd.read_csv('aapl_no_dates.csv')
df2.head(3)

daterange = pd.date_range(start='6/1/2017', end='6/30/17', freq='B') # freq=B  considers only business days
df2.set_index(daterange)
```

```python
# we can get date in many formats so we can convert it to standard format

dates = ['2017/1/5', '2017.01.05', '2017-01-05']
pd.to_datetime(dates)
```
