```python
df2 = pd.read_csv('weather2.csv')
df2
```

```python
# df2.pivot(index='date', columns='city')  This will give error because for each date-city pair
# there are multiple values. So we have to use pivot table there

df2.pivot_table(index='date', columns='city') # Here for values of each date-city
                                              # combination average functionality is used.
```

```python
# In above example instead of average we can use sum as values of each date-city combination
df2.pivot_table(index='date', columns='city', aggfunc='sum')
```
